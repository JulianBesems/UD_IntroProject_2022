# Dealing with missing data

```python
df2 = pd.DataFrame({'York':[1,5,np.nan],
                  'Adur':[5,np.nan,10],
                  'Birmingham':[1,5,3]})
```

```python
df2
```

```python
df2.dropna() #remove all NaN
```

```python
df2.fillna(value='FILL VALUE') # fill value
```

```python
df2['York'].fillna(value=df2['York'].mean())
```

Concatenation

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3']},
                        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7']},
                         index=[4, 5, 6, 7]) 
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

Pivot table

```python
mylist = {'A':['dog','dog','dog','cat','cat','cat'],
     'B':['one','one','two','two','one','one'],
       'C':['x','y','x','y','x','y'],
       'D':[1,3,2,5,4,1]}

df = pd.DataFrame(mylist)
```

```python
df.pivot_table(values='D',index=['A', 'B'],columns=['C'])
```

# Visualisation of data

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Basic Matplotlib commands

```python
#fig = plt.figure(figsize=(8,10), dpi=100)
plt.plot(df["Income Score"], df["Population aged 16-59"], 'r') # 'r' is the color red
plt.xlabel('Income Score')
plt.ylabel('Population')
plt.title('String Title Here')
plt.show()
```
