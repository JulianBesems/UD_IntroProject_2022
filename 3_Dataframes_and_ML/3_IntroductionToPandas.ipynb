# Dealing with missing data

```python
df2 = pd.DataFrame({'York':[1,5,None],
                  'Adur':[5,None,10],
                  'Birmingham':[1,5,3]})
```

```python
df2
```

```python
df2.dropna() #remove all NaN
```

```python
df2.fillna(value='FILL VALUE') # fill value
```

```python
df2['York'].fillna(value=df2['York'].mean(),inplace=True)
df2
```

-------
__None or np.nan__

Numpy is optimised for vector operations, based on numerical values. 
So to keep dataframes optimised, if you ever create an empty value that awaits filling, use np.nan instead of None 

(`None` is an `object` type, `np.nan` a `float`)

```python
None
np.nan
```

Creating Multiplots on Same Canvas

```python
a=[i/2 for i in range(10)]
a
```

```python
b=[i**2/4 for i in range(10)]
b
```
