**WHY SOM?**
According to the literature, SOM is a generic methodology, which has been applied in many classical modeling tasks such as **visualization** of a high dimensional space, **clustering and classification**, **prediction** and function approximation and many other tasks. 

There are different interpretations for SOM algorithm. 
One can compare SOM with other **clustering methods** or to compare it with space transformation and feature extraction methods such as Principal Component Analysis (PCA).

How are we going to analyze the relevances of cancalled airflight? How to use SOM?

Hypothesis:
* We can see the relavance by the distribution pattern of features.

Question:
* If we only show the features of data in the order of dataset, there would be only cloudy distribution. -> sort <- SOM

Steps:
* Train a SOM to get a general pattern/clusters of the dataset.
* Compare/visulize the features according to the SOM clusters.
 

## Select related features from the dataset, including internal causes and external causes

```python
# select the features that maybe related
df = df[["Month","DayofMonth", "DayOfWeek","DepTime", "AirTime",
         "Distance", "SecurityDelay","WeatherDelay", "NASDelay", "CarrierDelay",
         "ArrDelay", "DepDelay", "LateAircraftDelay", "Cancelled"]]
```

```python
# fill the NaN with 0
df = df.fillna(0)
```

```python
# select features for SOM to cluster
clustering_vars = ["Month", "DayofMonth", "DepTime", "AirTime", 
                   "LateAircraftDelay", "DepDelay", "ArrDelay", "CarrierDelay"]
```

```python
# read data
data = df[clustering_vars].values
```

```python
names = clustering_vars
names
```

```python
df.describe()
```

## train a SOM with features that we assume internal causes

```python
som = SOMFactory().build(data, mapsize=[20,20],
                            normalization = 'var', initialization='random', component_names=names, lattice="hexa")
```

```python
%%time
som.train(n_job=4, verbose=False, train_rough_len=3, train_finetune_len=10)
```

```python
# visulization
from sompy.visualization.mapview import View2D
view2D  = View2D(20,20,"", text_size=7)
view2D.show(som, col_sz=5, which_dim="all", denormalize=True)
plt.show()
```

Results trained with rough 10, fine 30

<img src="rough10-fine30.png" width="1200"/>

Why we only train the SOM with selected features?

## Sort data in a way according to BMU

```python
df.shape
```

```python
# addtion of bmu indexes to data frame
df["bmus"] = som.project_data(data)
df.shape
```

```python
df.columns
```

## The distribution pattern comparison of cancellation and other features.

```python
# Addition of some exogeneous features to the map that are not trained features, cancelled
exogeneous_vars = [c for c in df.columns if not c in clustering_vars+["Cancelled", "bmus"]] 
```

```python
df = df[clustering_vars + exogeneous_vars + ["Cancelled"] + ["bmus"]]
```

```python
# group by and mean value and fill with empty if there were no BMUs for a certain cell
map_size_total = som.codebook.mapsize[0]*som.codebook.mapsize[1]

empirical_codebookdf = df.groupby("bmus", as_index=True, dropna=True).mean()
for i in range(map_size_total):
    if i not in empirical_codebookdf.index:
        empirical_codebookdf.loc[i] = [np.NaN for _ in range(len(empirical_codebookdf.columns))]
empirical_codebookdf = empirical_codebookdf.sort_index()
empirical_codebook = empirical_codebookdf.values
```

```python
empirical_codebook.shape
```

```python
som.codebook.mapsize + [empirical_codebook.shape[-1]]
```

```python
matplotlib.rcParams.update({'font.size': 10})
plot_hex_map(empirical_codebook.reshape(som.codebook.mapsize + [empirical_codebook.shape[-1]]), 
             titles=df.columns[:-1], shape=[4, 5], colormap=None)
plt.show()
```

Conclusion:

* Cancellation is related to Month. Late months are more likely to have cancelled flight

* Flights with shorter distance have higher chance to be cancelled than long distance flights

* The delays occuring because of previous flight (security delay, arrival or departure delay) or National air system delay are not typical causes for flight cancellation.

## Optional version: Train more models and select the one with least error.

```python
%%time
# TBD
# Train the model with different parameters. The more, the better. Each iteration is stored in disk for further study
for i in range(5):
    som = SOMFactory().build(data, mapsize=[random.choice(list(range(15,25))),
                                            random.choice(list(range(15,25)))],
                            normalization = 'var', initialization='random', component_names=names,lattice="hexa")
    som.train(n_job=4, verbose=False, train_rough_len=3, train_finetune_len=10)
    joblib.dump(som, "./models/model_{}.joblib".format(i))
```

```python
models_pool = glob.glob("./models/model*")
errors=[]
for model_filepath in models_pool:
    sm = joblib.load(model_filepath)
    topographic_error = sm.calculate_topographic_error()
    quantization_error = sm.calculate_quantization_error()
    errors.append((topographic_error, quantization_error))
e_top, e_q = zip(*errors)
```

```python
plt.scatter(e_top, e_q)
for i in range(len(e_top)):
    plt.text(e_top[i], e_q[i], str(i))
plt.xlabel("Topographic error")
plt.ylabel("Quantization error")
plt.show()
```

```python
# select the model with very low topographic error and low quantization error
selected_model = 2
som = joblib.load(models_pool[selected_model])

topographic_error = sm.calculate_topographic_error()
quantization_error = sm.calculate_quantization_error()
print ("Topographic error = %s\n Quantization error = %s" % (topographic_error, quantization_error))
```

```python
# visulization
from sompy.visualization.mapview import View2D
view2D  = View2D(20,20,"", text_size=7)
view2D.show(som, col_sz=5, which_dim="all", denormalize=True)
plt.show()
```
